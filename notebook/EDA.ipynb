## Analysing which gender performed better who had lunch or not.

```python
plt.figure(figsize=(15,6))
plt.subplot(1,3,1)
sns.histplot(x='Average score',data=df,color='g',kde=True)


plt.subplot(132)
sns.histplot(x='Average score',hue='lunch',data=df[df['gender']=='male'],kde=True)

plt.subplot(133)
sns.histplot(x='Average score',hue='lunch',data=df[df['gender']=='female'],kde=True)
```
